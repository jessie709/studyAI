```python
import os
import pandas as pd
import numpy as np

x = os.listdir('./datas/ORL')
df = pd.DataFrame(x,columns=['file_name'])
df
```

```python
df['file_group_name'] = df.file_name.str.extract(r'(.*)(_)')[0]
df
```

```python
df['file_group_name'] = df.file_name.str.extract(r'(.*)(_)')[0]
grouped = df.groupby(['file_group_name'])
k=0.7
for v in grouped.groups.values():
    data = df['file_name'][v]
    train = data.sample(frac=k)
    test = data.drop(train.index)
    print(f'train is {train.values}')
    print(f'test is {test}')
    print(f'*'*15)
```
