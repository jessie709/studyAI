# 数据描述

- Table 1: 用户线下消费和优惠券领取行为
    - User_id: 用户 ID
    - Merchant_id: 商户 ID
    - Coupon_id: 优惠券 ID：null 表示无优惠券消费，此时 Discount_rate和 Date_received字段无意义
    - Discount_rate: 优惠率：$x \in [0,1]$ 代表折扣率；x:y 表示满x 减y。单位：元
    - Distance: user经常活动的地点离该 merchant 的最近门店距离是 $x \times 500$米
        -（如果是连锁店，则取最近的一家门店），$x \in [0,10]$；
        - null 表示无此信息，
        - 0表示低于500米，
        - 10表示大于5公里；
    - Date_received: 领取优惠券日期
    - Date: 消费日期：
        - 如果 Date=null & Coupon_id != null，该记录表示领取优惠券但没有使用，即负样本；
        - 如果 Date != null & Coupon_id = null，则表示普通消费日期；
        - 如果 Date != null & Coupon_id != null，则表示用优惠券消费日期，即正样本；




```python
import pandas as pd
import numpy as np

table1_file_path = './ccf_offline_stage1_train.csv'
#table1_file_path = './test.csv'
table1_data = pd.read_csv(table1_file_path, parse_dates=[5,6], dtype={'Coupon_id': 'str'})

df1 = table1_data.dropna(axis=0)

df1['Action'] = 3
# 满减 0
# 打折 1
#df['Discount_type'] = np.where(df['Discount_rate'].str.contains(':', na=False), 0, 1)
#new = df['Discount_rate'].str.split(':', n=1, expand=True)
#df['Discount_new_rate'] = (new[0].astype(float) - new[1].astype(float)) / new[0].astype(float)
df1['Distance_date_for_consume'] = (df1['Date'] - df1['Date_received']).dt.days
df1['Target'] = np.where(df1['Distance_date_for_consume'] <= 15, 1, 0)
df1
```

- Table2:用户上线点击/消费和优惠劵领取行为
    - User_id:用户ID
    - Merchant_id:商户ID
    - Action : 0点击，1 购买，2领取优惠劵
    - Coupon_id: 优惠劵ID: null表示无优惠劵，此时的Discount_rate和Date_received字段无意义。“fixed”表示该交易是限时低活动。
    - Discount_rate: 优惠率：
        - $x \in [0,1]$ 代表折扣率;
        - x:y表示满x减y;
        - "fixed"低价限时优惠；
    - Date_received: 领取优惠劵日期；
    - Date: 消费日期：
        - 如果Date=null&Coupon_id!=null,该记录表示领取优惠劵但没有使用；
        - 如果Date!=null&Coupon_id=null,则表示普通消费日期；
        - 如果Date!=null&Coupon_id!=null,则表示用优惠劵消费日期；

```python
import pandas as pd
import numpy as np

table2_file_path = './ccf_online_stage1_train.csv'
table2_data = pd.read_csv(table2_file_path, parse_dates=[5,6], dtype={'Coupon_id':'str'})

df2 = table2_data.dropna(axis=0)

df2['Distance'] = 0
df2['Distance_date_for_consume'] = (df2['Date'] - df2['Date_received']).dt.days
df2['Target'] = np.where(df2['Distance_date_for_consume'] <= 15, 1, 0)
df2
```

```python
# concate df1 and df2
import pandas as pd

frames = [df1, df2]

df = pd.concat(frames, sort=False,ignore_index=True, axis=0)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['Coupon_id'])
df['Coupon_id'] = le.transform(df['Coupon_id'])

le.fit(df['Discount_rate'])
df['Discount_rate'] = le.transform(df['Discount_rate'])

df
```

```python
# model
y = df['Target']
X = df[['User_id', 'Merchant_id', 'Coupon_id', 'Discount_rate', 'Distance', 'Action']]

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(X, y)
```

- Table3: 用户O2O线下优惠券使用预测样本
    - User_id: 用户 ID
    - Merchant_id: 商户 ID

    - Discount_rate: 优惠率：$x \in [0,1]$代表折扣率；x:y 表示满x 减y。单位：元

            - 10表示大于5公里；

```python
import pandas as pd

table3_file_path = './ccf_offline_stage1_test_revised.csv'
table3_data = pd.read_csv(table3_file_path)

df3 = table3_data

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(df3['Discount_rate'])
df3['Discount_rate'] = le.transform(df3['Discount_rate'])

from sklearn.impute import SimpleImputer
imp = SimpleImputer(strategy='constant', fill_value=0)
df31 = pd.DataFrame(imp.fit_transform(df3), columns=df3.columns)

predict_proba = rfc.predict_proba(df31)
#predict = rfc.predict(df31)
#predict
```
