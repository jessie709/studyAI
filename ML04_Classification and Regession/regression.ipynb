# metrics

## Explained_variance_score

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
%matplotlib inline
```

```python
# Explained_variance_score
from sklearn.metrics import explained_variance_score

y_true = [3, -0.5, 2, 7] # ground truth
y_pred = [2.5, 0.0, 2, 8] # prediction
print('explained_variance_score: {}'.format(explained_variance_score(y_true, y_pred)))

y_true = [[0.5, 1], [-1, 1], [7, -6]]
y_pred = [[0, 2], [-1, 2], [8, -5]]
print('explained_variance_score: {}'.format(explained_variance_score(y_true, y_pred, multioutput='raw_values')))
print('explained_variance_score: {}'.format(explained_variance_score(y_true, y_pred, multioutput=[0.3, 0.7])))
```

## Mean absolute error

```python
# Mean absolute error
from sklearn.metrics import mean_absolute_error

y_true = [3, -0.5, 2, 7] # ground truth
y_pred = [2.5, 0.0, 2, 8] # prediction
print('mean_absolute_error: {}'.format(mean_absolute_error(y_true, y_pred)))

y_true = [[0.5, 1], [-1, 1], [7, -6]]
y_pred = [[0, 2], [-1, 2], [8, -5]]
print('mean_absolute_error: {}'.format(mean_absolute_error(y_true, y_pred)))
print('mean_absolute_error: {}'.format(mean_absolute_error(y_true, y_pred, multioutput='raw_values')))
print('mean_absolute_error: {}'.format(mean_absolute_error(y_true, y_pred, multioutput=[0.3, 0.7])))
```

## Mean squared error

```python
# Mean squared error
from sklearn.metrics import mean_squared_error

y_true = [3, -0.5, 2, 7] # ground truth
y_pred = [2.5, 0.0, 2, 8] # prediction
print('mean_squared_error: {}'.format(mean_squared_error(y_true, y_pred)))

y_true = [[0.5, 1], [-1, 1], [7, -6]]
y_pred = [[0, 2], [-1, 2], [8, -5]]
print('mean_squared_error: {}'.format(mean_squared_error(y_true, y_pred)))
```

## Mean squared logarithmic error

```python
# Mean squared logarithmic error
from sklearn.metrics import mean_squared_log_error

y_true = [3, 5, 2.5, 7] # ground truth
y_pred = [2.5, 5, 4, 8] # prediction
print('mean_squared_log_error: {}'.format(mean_squared_log_error(y_true, y_pred)))

y_true = [[0.5, 1], [1, 2], [7, 6]]
y_pred = [[0.5, 2], [1, 2.5], [8, 8]]
print('mean_squared_log_error: {}'.format(mean_squared_log_error(y_true, y_pred))) 
```

## Median absolute error

```python
# Median absolute error
from sklearn.metrics import median_absolute_error

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
print('median_absolute_error: {}'.format(median_absolute_error(y_true, y_pred)))
```

```python
# RÂ² score
from sklearn.metrics import r2_score

y_true = [3, -0.5, 2, 7]
y_pred = [2.5, 0.0, 2, 8]
print(r2_score(y_true, y_pred)) 

y_true = [[0.5, 1], [-1, 1], [7, -6]]
y_pred = [[0, 2], [-1, 2], [8, -5]]
print(r2_score(y_true, y_pred, multioutput='variance_weighted'))
print(r2_score(y_true, y_pred, multioutput='uniform_average'))
print(r2_score(y_true, y_pred, multioutput='raw_values'))
print(r2_score(y_true, y_pred, multioutput=[0.3, 0.7]))
```

# Regreesion algorithms

## KNN

```python
from sklearn.neighbors import KNeighborsRegressor

X = np.linspace(0, 100, 100).reshape(-1, 1)
y = np.zeros(X.shape)
y[50:, 0] = 1

reg_knn = KNeighborsRegressor(n_neighbors=2)
reg_knn.fit(X, y) 
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.scatter(X, y)
plt.title('Row Data')

plt.subplot(122)
plt.scatter(X, y)
k = 1
for i in range(k):
    test_sample = np.random.uniform(X[49], X[50])
    plt.scatter(test_sample, reg_knn.predict([test_sample]))
plt.title('Regression KNN')
    
```

## Radius-KNN

```python
from sklearn.neighbors import RadiusNeighborsRegressor

reg_knn_radius = RadiusNeighborsRegressor(radius=1.0)
reg_knn_radius.fit(X, y) 
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.scatter(X, y)
plt.title('Row Data')

plt.subplot(122)
plt.scatter(X, y)
k = 1
for i in range(k):
    test_sample = np.random.uniform(X[49], X[50])
    plt.scatter(test_sample, reg_knn_radius.predict([test_sample]))
plt.title('Regression Radius-KNN')
```

## SVM

```python
import numpy as np
from sklearn.svm import SVR

# Generate sample data
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel() # 1-d
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y)
plt.title('Row Data')
```

```python
# Add noise to targets
y[::5] += 3 * (0.5 - np.random.rand(8))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X, y)
plt.title('Data with noise')
```

```python

# Fit regression model
svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_lin = SVR(kernel='linear', C=1e3)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)
y_rbf = svr_rbf.fit(X, y).predict(X)
y_lin = svr_lin.fit(X, y).predict(X)
y_poly = svr_poly.fit(X, y).predict(X)
```

```python
# Look at the results
lw = 2
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='darkorange', label='data')
plt.plot(X, y_rbf, color='navy', lw=lw, label='RBF model')
plt.plot(X, y_lin, color='c', lw=lw, label='Linear model')
plt.plot(X, y_poly, color='cornflowerblue', lw=lw, label='Polynomial model')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
# y = 1 * x_0 + 2 * x_1 + 3
y = np.dot(X, np.array([1, 2])) + 3
reg = LinearRegression().fit(X, y)
print(reg.score(X, y))
print(reg.coef_)
print(reg.intercept_) 
print(reg.predict(np.array([[3, 5]])))
```

```python
from mpl_toolkits.mplot3d import Axes3D

xx = np.arange(0, 10, 0.25)
yy = np.arange(0, 10, 0.25)
xx,yy = np.meshgrid(xx, yy)
z = 1 * xx + 2 * yy + 3
fig = plt.figure(figsize=(12, 6))

ax = fig.add_subplot(121, projection='3d', title='Row Data')
ax.scatter(X[:,0], X[:,1], y, s=100, c='r', marker='o')

ax = fig.add_subplot(122, projection='3d', title='Regression-LR Surface')
ax.plot_surface(xx, yy, z, color='g')
```

## Bayesian

```python
from sklearn.linear_model import BayesianRidge
```

```python
# Generating simulated data with Gaussian noise
np.random.seed(0)
X = np.linspace(0, 10, 100)
noise = np.random.normal(0, 1, len(X))
y = np.sqrt(X) * np.sin(X) + 0.1 * noise
X.shape, y.shape
```

```python
fig = plt.figure()
plt.scatter(X, y, color='gold', marker='^')
plt.title('Trainning Data')
```

```python
# Fit the Bayesian Ridge Regression
reg_bayes = BayesianRidge(compute_score=True)
reg_bayes.fit(np.vander(X, 10), y)
```

```python
# testing data
X_plot = np.arange(0, 10, 0.5)
y_plot = np.sqrt(X_plot) * np.sin(X_plot)
y_predict = reg_bayes.predict(np.vander(X_plot, 10))
```

```python
fig = plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.title('Regression Bayesian')
plt.scatter(X, y, color='gold', marker='^', label='Train Data')
plt.plot(X_plot, y_predict, label='Bayesian Ridge')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

plt.subplot(122)
coefs = reg_bayes.coef_
plt.title("Histogram of the weights")
plt.hist(coefs, bins=len(coefs), log=True)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
```

```python
fig = plt.figure(figsize=(8, 6))
plt.scatter(X, y)
plt.title('Row Data')
```

```python
y[::5] += 3 * (0.5 - rng.rand(16))
```

```python
fig = plt.figure(figsize=(8, 6))
plt.scatter(X, y)
plt.title('Data with noise')
```

```python
# Fit regression model
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(X, y)
regr_2.fit(X, y)
```

```python
# Predict
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
```

```python
# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, s=20, edgecolor="black",
            c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",
         label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=1000, n_features=4, n_informative=2, 
                       random_state=0, shuffle=True)
regr = RandomForestRegressor(max_depth=5, random_state=0, n_estimators=100)
regr.fit(X[:800], y[:800])

print(regr.feature_importances_)
print(regr.score(X[800:], y[800:])) # R^2 
```
