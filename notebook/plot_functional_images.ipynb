# Plot functional images

绘制各类函数的可视化图，用于理解函数的特性

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

print(f'numpy version is {np.__version__}')
print(f'matplotlib version is {matplotlib.__version__}')
```

## $y=a+bx+cx^2$

```python
a = 5
b = 6.89
c = 100
N = 500
x = np.random.rand(N)
y = a + b * x + c * x ** 2

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(x, y, 'o')
axes[0].legend([r'$c$ is positive'])

c = -100
y = a + b * x + c * x ** 2
axes[1].plot(x, y, '+')
axes[1].legend([r'$c$ is negative'])
```

## $y=\sqrt{1-x^2}$

```python

N = 500
x = np.random.rand(N)
y = np.sqrt(1 - x ** 2)
x = np.concatenate((x, x, -x, -x))
y = np.concatenate((y, -y, y, -y))

fig, axes = plt.subplots()
axes.plot(x, y, 'o')
```
