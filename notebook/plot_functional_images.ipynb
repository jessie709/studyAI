# Plot functional images

绘制各类函数的可视化图，用于理解函数的特性

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

print(f'numpy version is {np.__version__}')
print(f'matplotlib version is {matplotlib.__version__}')
```

## $y=a+bx+cx^2$

```python
a = 5
b = 6.89
c = 100
N = 500
x = np.random.rand(N)
y = a + b * x + c * x ** 2

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].plot(x, y, '+')
axes[0].legend([r'$c$ is positive'])

c = -100
y = a + b * x + c * x ** 2
axes[1].plot(x, y, '.')
axes[1].legend([r'$c$ is negative'])
```

## $y=\sqrt{1-x^2}$

```python
N = 500
x = np.random.rand(N)
y = np.sqrt(1 - x ** 2)
x = np.concatenate((x, x, -x, -x))
y = np.concatenate((y, -y, y, -y))

fig, axes = plt.subplots()
axes.plot(x, y, 'o')
```

## $y=\theta_0+\theta_1 x_1+\theta_2 x_2$

```python
t0, t1, t2 = 8.5, 9.8, 112.5
N = 50
X = np.random.rand(N)
Y = np.random.rand(N)
X, Y = np.meshgrid(X, Y)
Z = t0 + t1 * X + t2 * Y

# plot 3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```

## $y=\theta_0+\theta_1 x_1+\theta_2 x_2+\theta_3 xy$

```python
t0, t1, t2, t3 = 8.5,99.8, 112.5, 888
N = 50
X = np.random.rand(N)
Y = np.random.rand(N)
X, Y = np.meshgrid(X, Y)
Z = t0 + t1 * X + t2 * Y + t3 * X * Y

# plot 3D
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
```
