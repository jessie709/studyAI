```python
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
```

```python
mat = scipy.io.loadmat('../data/andrew/ex3data1.mat')
X = mat['X']
mat
```

```python
mat['y'][0, :]
```

```python
from matplotlib import gridspec

rows, cols = 10, 10

# fig = plt.figure(figsize=(rows+1, cols+1))

gs = gridspec.GridSpec(rows, cols,
         wspace=0.0, hspace=0.0, 
         top=1.-0.5/(rows+1), bottom=0.5/(rows+1), 
         left=0.5/(cols+1), right=1-0.5/(cols+1)) 

n = rows * cols
random_idx = np.random.choice(X.shape[0], n, replace=False)
sample = X[random_idx, :]

cur_idx = 0
for i in range(rows):
    for j in range(cols):
        s = sample[cur_idx, :]
        s = np.reshape(s, (20, 20))
        ax = plt.subplot(gs[i, j])
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax.imshow(s, cmap='gray')
        cur_idx += 1
```
