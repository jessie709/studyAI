## Start simple -- univariate scalar optimization

$$
f(x)=\sin(x)e^{\left[-0.1(x-0.6)^2\right]}
$$

plotted between $x=-10$ to $x=10$

```python
import numpy as np
import matplotlib.pyplot as plt

def scalar1(x):
    return np.sin(x) * np.exp(-0.1 * (x - 0.6) ** 2)

x = np.linspace(-10, 10, 100)
y = scalar1(x)

plt.plot(x, y)
```

```python
from scipy import optimize

result = optimize.minimize_scalar(scalar1)
print(result)

x = np.linspace(-10, 10, 100)
y = scalar1(x)

plt.axvline(x=result['x'], c='r', ls='--')
plt.plot(x, y)
```

### What if the variable are bounded?

```python
result = optimize.minimize_scalar(scalar1, bounds=(0, 10), method='Bounded')
print(result)

x = np.linspace(-10, 10, 100)
y = scalar1(x)

plt.axvline(x=result['x'], c='r', ls='--')
plt.plot(x, y)
```
