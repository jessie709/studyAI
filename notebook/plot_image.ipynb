```python
X = mat['X']
X.shape
```

```python
rows, cols = 2, 2

fig, axes = plt.subplots(rows, cols)

n = rows * cols
random_idx = np.random.choice(X.shape[0], n, replace=False)
sample = X[random_idx, :]
current_row = -1
for i in range(sample.shape[0]):
    current_col = i % cols
    if current_col == 0:
        current_row += 1
    s = sample[i, :]
    s = np.reshape(s, (20, 20))
    axis = axes[current_row][current_col]
    axis.get_xaxis().set_visible(False)
    axis.get_yaxis().set_visible(False)
    axis.imshow(s, cmap='gray')
```

```python
from matplotlib import gridspec

rows, cols = 10, 10

# fig = plt.figure(figsize=(rows+1, cols+1))

gs = gridspec.GridSpec(rows, cols,
         wspace=0.0, hspace=0.0, 
         top=1.-0.5/(rows+1), bottom=0.5/(rows+1), 
         left=0.5/(cols+1), right=1-0.5/(cols+1)) 

n = rows * cols
random_idx = np.random.choice(X.shape[0], n, replace=False)
sample = X[random_idx, :]

cur_idx = 0
for i in range(rows):
    for j in range(cols):
        s = sample[cur_idx, :]
        s = np.reshape(s, (20, 20))
        ax = plt.subplot(gs[i, j])
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
        ax.imshow(s, cmap='gray')
        cur_idx += 1
```
